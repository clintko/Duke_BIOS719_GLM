# Analysis 1
**Consider the respiratory status at visit 1 as the outcome. Fit an appropriate model evaluating the treatment effect with a possible adjustment for gender and age. Interpret the results.**

```r
###
set.seed(0)

### combine
dat = dat_raw %>% 
    group_by(trt, age, gender) %>% 
    summarize(y = sum(visit1), n = n()) %>% 
    as.data.frame 
#dat_n1 = dat_raw %>% select(visit1, treatment, age) %>% 

### display the six rows
dat %>% sample_n(size = 6)
```

## fit logit

```r
fit_logit_ln     = glm(cbind(y, n-y) ~ trt + age + gender,            data = dat, family = binomial(link = "logit"), x = T)
fit_logit_sq     = glm(cbind(y, n-y) ~ trt + age + gender + I(age^2), data = dat, family = binomial(link = "logit"), x = T)
fit_logit_ln_int = glm(cbind(y, n-y) ~ trt + age + gender + trt*age,  data = dat, family = binomial(link = "logit"), x = T)
```

```r
###
df = cbind(dat, f.glm.diag(fit_logit_ln))
gp  = ggplot(df, aes(x = age, y = sdres)) + facet_wrap(~trt)
gp1  = gp +
    geom_point() + 
    geom_hline(yintercept = 0, lty = 2) +
    geom_smooth(method = "loess") +
    labs(title = "cbind(y, n-y) ~ trt + age",
         x = "age", y = "standardized\ndeviane residual")

###
df = cbind(dat, f.glm.diag(fit_logit_sq))
gp  = ggplot(df, aes(x = age, y = sdres)) + facet_wrap(~trt)
gp2  = gp +
    geom_point() + 
    geom_hline(yintercept = 0, lty = 2) +
    geom_smooth(method = "loess") +
    labs(title = "cbind(y, n-y) ~ trt + age + age^2",
         x = "age", y = "standardized\ndeviane residual")

###
df = cbind(dat, f.glm.diag(fit_logit_ln_int))
gp  = ggplot(df, aes(x = age, y = sdres)) + facet_wrap(~trt)
gp3  = gp +
    geom_point() + 
    geom_hline(yintercept = 0, lty = 2) +
    geom_smooth(method = "loess") +
    labs(title = "cbind(y, n-y) ~ trt + age + trt * age",
         x = "age", y = "standardized\ndeviane residual")


options(repr.plot.width = 5, repr.plot.height = 8)
grid.arrange(gp1, gp2, gp3, nrow = 3, ncol = 1)
```

```r
fit_linear = fit_logit_ln$fitted.values
fit_square = fit_logit_sq$fitted.values
fit_ln_int = fit_logit_ln_int$fitted.values


gp = ggplot(dat, aes(x = age)) + 
    facet_wrap(~trt) + 
    geom_point(aes(y = y / n)) +
    geom_line( aes(y = fit_linear, color = 'fit linear')) +
    geom_line( aes(y = fit_square, color = 'fit square')) +
    geom_line( aes(y = fit_ln_int, color = 'fit linear interaction')) +
    scale_colour_manual("", 
                      breaks = c("fit linear", "fit square", 'fit linear interaction'),
                      values = c("steelblue",  "salmon", "green")) +
    labs(y = "Predicted Probability\nof Success",
         x = "age")

options(repr.plot.width = 6, repr.plot.height = 3)
print(gp)
```

# fit cloglog

```r
fit_clglg_ln     = glm(cbind(y, n-y) ~ trt + age,            data = dat, family = binomial(link = "cloglog"), x = T)
fit_clglg_sq     = glm(cbind(y, n-y) ~ trt + age + I(age^2), data = dat, family = binomial(link = "cloglog"), x = T)
fit_clglg_ln_int = glm(cbind(y, n-y) ~ trt + age + trt*age,  data = dat, family = binomial(link = "cloglog"), x = T)
```

```r
###
df = cbind(dat, f.glm.diag(fit_clglg_ln))
gp  = ggplot(df, aes(x = age, y = sdres)) + facet_wrap(~trt)
gp1  = gp +
    geom_point() + 
    geom_hline(yintercept = 0, lty = 2) +
    geom_smooth(method = "loess") +
    labs(title = "cbind(y, n-y) ~ trt + age",
         x = "age", y = "standardized\ndeviane residual")

###
df = cbind(dat, f.glm.diag(fit_clglg_sq))
gp  = ggplot(df, aes(x = age, y = sdres)) + facet_wrap(~trt)
gp2  = gp +
    geom_point() + 
    geom_hline(yintercept = 0, lty = 2) +
    geom_smooth(method = "loess") +
    labs(title = "cbind(y, n-y) ~ trt + age + age^2",
         x = "age", y = "standardized\ndeviane residual")

###
df = cbind(dat, f.glm.diag(fit_clglg_ln_int))
gp  = ggplot(df, aes(x = age, y = sdres)) + facet_wrap(~trt)
gp3  = gp +
    geom_point() + 
    geom_hline(yintercept = 0, lty = 2) +
    geom_smooth(method = "loess") +
    labs(title = "cbind(y, n-y) ~ trt + age + trt * age",
         x = "age", y = "standardized\ndeviane residual")


options(repr.plot.width = 5, repr.plot.height = 8)
grid.arrange(gp1, gp2, gp3, nrow = 3, ncol = 1)
```

```r
fit_linear = fit_clglg_ln$fitted.values
fit_square = fit_clglg_sq$fitted.values
fit_ln_int = fit_clglg_ln_int$fitted.values


gp = ggplot(dat, aes(x = age)) + 
    facet_wrap(~trt) + 
    geom_point(aes(y = y / n)) +
    geom_line( aes(y = fit_linear, color = 'fit linear')) +
    geom_line( aes(y = fit_square, color = 'fit square')) +
    geom_line( aes(y = fit_ln_int, color = 'fit linear interaction')) +
    scale_colour_manual("", 
                      breaks = c("fit linear", "fit square", 'fit linear interaction'),
                      values = c("steelblue",  "salmon", "green")) +
    labs(y = "Predicted Probability\nof Success",
         x = "age")

options(repr.plot.width = 6, repr.plot.height = 3)
print(gp)
```

# Analysis 3
**Consider a model utilizing the longitudinal measurements of the
respiratory status at all four visits. Account for a possible correlation of visit
outcomes within a patient. Additionally, estimate and interpret impact of time on
the treatment effect. Interpret the final model results**

```r
dat_clust = dat_raw %>% select(patient, starts_with("visit")) %>% gather(visit, value, -patient)
dat_clust$patient = as.character(dat_clust$patient)
head(dat_clust)
```

```r
dat = dat_clust %>% filter(patient %in% c(50, 100, 150, 200, 250, 300, 350, 400, 500))

gp = ggplot(dat, aes(x = visit, y = value, group = patient)) + 
    geom_line() + 
    facet_wrap(~patient)

options(repr.plot.height = 4, repr.plot.width = 6)
print(gp)
```
